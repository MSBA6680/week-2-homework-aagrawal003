```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt

# Fetch AirBNB data from Yahoo Finance for 10 years
df = yf.download("ABNB", start="2015-01-01", end="2025-04-04")

# Calculate daily returns and annualized volatility
df['Returns'] = df['Close'].pct_change()
df['Volatility'] = df['Returns'].rolling(window=252).std() * np.sqrt(252)

# Plot the AirBnb prices
plt.figure(figsize=(14, 10))
plt.subplot(3, 1, 1)
plt.plot(df.index, df['Close'], label='AirBNB Prices', color='blue')
plt.title('AirBNB Prices')
plt.ylabel('Price')
plt.legend()


# Plot the stock return
plt.subplot(3, 1, 2)
plt.plot(df.index, df['Returns'], label='Stock Return', color='Green')
plt.title('Stock Returns')
plt.ylabel('Stock Return')
plt.xlabel('Year')
plt.legend()


# Plot the annualized volatility
plt.subplot(3, 1,3)
plt.plot(df.index, df['Volatility'], label='Annualized Volatility', color='red')
plt.title('AirBNB Annualized Volatility')
plt.ylabel('Annualized Volatility')
plt.xlabel('Year')
plt.legend()

plt.tight_layout()
plt.show()
```
